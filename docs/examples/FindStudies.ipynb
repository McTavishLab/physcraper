# Finding studies for a `Physcraper` run

```python
from opentree import OT
```

```python
OT.get_ottid_from_name("Malvaceae")
```

```python
phylesystem_studies_resp = OT.find_trees("279960", search_property ='ot:ottId')
```

```python
phylesystem_only_studies = set()
for study in phylesystem_studies_resp.response_dict['matched_studies']:
    study_id = study['ot:studyId']
    phylesystem_only_studies.add(study_id)
```

```python
cites_phyl = OT.get_citations(phylesystem_only_studies)
```
